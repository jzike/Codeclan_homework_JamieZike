### 2. We're going to be looking at the number of Calories in each drink. Calculate some quick summary statistics (e.g. the mean, the maximum and minimum).

```python
#get summary stats for Calories variable
drinks_content.Calories.describe()
```

### 3. Select the variables Beverage_category, Beverage, Beverage prep and Calories from the drinks_content data frame, and assign the selected columns to a new data frame called drinks. Check if there are any missing values anywhere in the data, and drop any rows containing them if there are.

```python
#create new dataset subset
drinks = (
    drinks_content
    .loc[:, ['Beverage_category', 'Beverage', 'Beverage_prep', 'Calories']]
    .copy()
)
```

```python
drinks.isna().sum()
#no nas
```

### 4. In your drinks dataframe create a new variable (column) called calorie_diff, which stores the difference between 135 calories (135 calories = 10 cubes of sugar!) and the calories in each drink. 

```python
drinks = (
    drinks
    .assign(
    calorie_diff = (135 - drinks.Calories)
    )
    .copy()
)
```

### 5. Summarise the mean number of calories in each beverage category.

```python
(
    drinks
    .groupby('Beverage_category')
    .Calories
    .mean()
    .reset_index(name = 'mean_calories')
)
```

### 6. Which Beverage_prep type contains more than the average calories of all drinks?

```python
# make the grouped dataset
beverage_prep_grouped = (
    drinks
    .groupby('Beverage_prep')
    .Calories
    .mean()
    .reset_index(name = 'mean_calories')
    .copy()
)
```

```python
#filter grouped dataset by mean Calories of main drinks dataset
beverage_prep_grouped.loc[(beverage_prep_grouped.mean_calories >= drinks.Calories.mean()), ['Beverage_prep', 'mean_calories']]
```

### 7. Which is the best type of coffee drink to get if you’re worried about consuming too many calories?

```python
drinks.Beverage_category.unique()
```

```python
coffee_drinks = ['Coffee', 'Frappuccino® Blended Coffee', 'Frappuccino® Light Blended Coffee', 'Classic Espresso Drinks', 'Signature Espresso Drinks']
```

```python
(
    drinks.loc[drinks.Beverage_category.isin(coffee_drinks)]
    .sort_values('Calories')
    .head(20)
)
#if watching calories, beverages from the Coffee category are the best, followed by Classic Espresso drinks
```

### 8. Filter the dataset drinks so we only take "Classic Espresso Drinks", and save this in a new data frame called espresso_drinks.

```python
espresso_drinks = (
    drinks
    .loc[(drinks.Beverage_category == 'Classic Espresso Drinks'), ]
    .copy()
)
```

### 9. Group your espresso_drinks data frame by the type of Beverage prep, and then find out the mean calories for each drink group.



```python
(
    espresso_drinks
    .groupby('Beverage_prep')
    .Calories
    .mean()
    .reset_index(name = 'mean_calories')
    .sort_values('mean_calories')
)
```

### 10. Get the same grouped mean Calories values as above, but this time sorted in descending order. 

```python
(
    espresso_drinks
    .groupby('Beverage_prep')
    .Calories
    .mean()
    .reset_index(name = 'mean_calories')
    .sort_values('mean_calories', ascending=False)
)
```
